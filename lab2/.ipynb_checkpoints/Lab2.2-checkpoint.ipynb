# Задача 2.2
$f(x) = 10^{-\sqrt{x}}- \sin(\pi \sqrt{x})- 0.9$

$f'(x) = -\frac{1}{2 \sqrt{x}} \left( \ln{(10)} \cdot 10 ^{-\sqrt{x}} + \pi \cos(\pi \sqrt{x})\right)$

## Решение

```python
import numpy as np
from numpy import abs, sqrt, sin, cos, pi, log
from matplotlib import pyplot as plt 
```

### Поиск отрезков локализации

```python
f = lambda x: 10**(-sqrt(x)) - sin(pi * sqrt(x)) - 0.9
df = lambda x: -1/(2 * sqrt(x)) * (log(10) * 10 ** (-sqrt(x)) + pi * cos(pi * sqrt(x)))
seg = [0, 3]
eps = 10**(-12)
```

```python
x = np.arange(*seg, 0.0001)
loc = np.array([[0, 0.001], [1.6, 1.9], [2.6, 2.9]])
m1 = np.array([df(loc[0][1]), df(loc[1][0]), df(loc[2][1])])
M1 = np.array([df(loc[0][0]), df(loc[1][1]), df(loc[2][0])])
[plt.scatter(x, f(x)) for x in loc]
plt.plot(x, f(x), "k", label="f(x)")
off = 1500
plt.plot(x[off:], df(x[off:]))
plt.grid()
plt.title("$y = f(x)$")
plt.savefig("report/221.png")
plt.savefig("report/221.eps")
```

### Метод Ньютона

```python
def newton(x0, func, dfunc, eps):
    x1 = x0 - func(x0) / dfunc(x0)
    it = 1
    while abs(x1 - x0) > eps:
        x0 = x1
        x1 = x0 - func(x0) / dfunc(x0)
        # print(f"\t{x1}")
        it += 1
    print(f"\tВыполнено {it} итераций. x = {np.round(x1, 12)}")
    return x1


x0 = (loc[:, 1] + loc[:, 0])/2
print('Поиск корней методом Ньютона.')
roots = np.array([newton(start, f, df, eps) for start in x0])
plt.scatter(roots, f(roots))
plt.plot(x, f(x), "k")
plt.grid()
plt.title("Корни уравнения ")
```

### Метод простых итераций

```python
def MPI(x0, m1, M1, f, eps):
    alpha = 2 / (m1 + M1)
    q = abs((M1-m1)/ (M1 + m1))
    x1 = x0 - alpha * f(x0)
    it = 1
    while abs(x1 - x0) > (1 - q) * eps / q:
        x0 = x1
        x1 = x0 - alpha * f(x0)
        it += 1
    
    print(f"\tВыполнено {it} итераций. x = {np.round(x1, 12)}")
    return x1
print('Поиск корней методом простых итераций.')
mpi_roots = np.array([MPI(start, m , M, f, eps) for start, m, M in zip(x0, m1, M1)])
plt.scatter(mpi_roots, f(mpi_roots), color='r')
plt.plot(x, f(x), "k")
plt.grid()
```

### Модификация методов

```python
def newton2(x0, func, dfunc, n, eps):
    x1 = x0 - func(x0) / dfunc(x0)
    r = [abs(f(x1))]
    it = 1
    while it < n:
        x0 = x1
        x1 = x0 - func(x0) / dfunc(x0)
        r.append(abs(f(x1)))
        it += 1
    return x1, np.array(r)

def MPI2(x0, m1, M1, f, n, eps):
    alpha = 2 / (m1 + M1)
    q = abs((M1-m1)/ (M1 + m1))
    x1 = x0 - alpha * f(x0)
    r = [abs(f(x1))]
    it = 1
    while it < n:
        x0 = x1
        x1 = x0 - alpha * f(x0)
        r.append(abs(f(x1)))
        it += 1
    return x1, np.array(r)

i = 0
for st, m, M in zip(x0, m1, M1):
    root, r = newton2(st, f, df, 10, eps)
    root2, r2 = MPI2(st, m, M, f, 10, eps)
    print(root, root2)
    n = np.arange(1, 11, 1)
    plt.plot(n, r, label="Невязка метода Ньютона")
    plt.plot(n, r2, label="Невязка метода простых итераций")
    plt.title(f"$x_0 = {st}$")
    plt.legend()
    plt.yscale('log')
    plt.savefig(f'report/222_{i}.png')
    plt.savefig(f'report/222_{i}.eps')
    plt.show()

    i += 1
```
