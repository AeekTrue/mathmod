## Задача 5.1


$
f(x) = \sqrt{x - 2},\ x \in [0, 4].
$

$\varepsilon = 10^{-6}. $

```python
n = 20
a, b = 2+0.2, 4
X_train = np.linspace(a, b, n+1)
f = lambda x: np.sqrt(x - 2)
Y_train = f(X_train)
X = np.linspace(a, b, 1000)
```

```python
def lagrange_polynomial(x, y):
    n = x.size
    delta_x = np.zeros((n, n))
    for i in range(n):
        for j in range(i):
            delta_x[i, j] = x[i] - x[j]
            delta_x[j, i] = -delta_x[i, j]

    def L(t):
        result = 0
        for i in range(n):
            # print(y[i])
            li = y[i]
            for j in range(n):
                if j == i:
                    continue
                li *= (t - x[j]) / delta_x[i, j]
            result += li
        return result
    return L    
```

```python
L = lagrange_polynomial(X_train, Y_train)
```

```python
Y_lag = np.array([L(x) for x in X])
Y = f(X)
```

```python
plt.plot(X, Y)
plt.plot(X, Y_lag)
```

```python
plt.plot(X, np.abs(Y - Y_lag), color='red', label='$|f(t) - L(t)|$')
plt.legend()
```
